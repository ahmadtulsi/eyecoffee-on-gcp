# Preprocess The Data

## RoCoLe dataset

```python
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/EyeCoffee_TrainingData/rocoleoriginal/Annotations/RoCoLe-classes.xlsx", 
                   usecols =['File', 'Multiclass.Label'])
df.head()
```

```python
df.rename(columns={"Multiclass.Label":"Label"}, inplace=True)
```

```python
file_path_rocole = "/content/drive/MyDrive/EyeCoffee_TrainingData/rocoleoriginal/Photos"
df["Filepath"] = file_path_rocole + "/" + df["File"]
df.tail()
```

```python
df['Label'] = df['Label'].replace(['rust_level_1', "rust_level_2", "rust_level_3", "rust_level_4"], 'rust')
```

## Rust and Leaf Miner dataset

```python
import fnmatch
import os

rust_file = []
for file in os.listdir('/content/drive/MyDrive/EyeCoffee_TrainingData/coffee-leaves-disease/rust_xml_image'):
    if fnmatch.fnmatch(file, '*.jpg'):
        rust_file.append(file)
```

```python
rust_dict = {'File': rust_file, 'Label': 'rust'}
df_rust = pd.DataFrame(rust_dict)
```

```python
file_path_rust= "/content/drive/MyDrive/EyeCoffee_TrainingData/coffee-leaves-disease/rust_xml_image"
df_rust["Filepath"] = file_path_rust + "/" + df_rust["File"]
df_rust.head()
```

```python
miner_file = []
for file in os.listdir('/content/drive/MyDrive/EyeCoffee_TrainingData/coffee-leaves-disease/miner_img_xml'):
    if fnmatch.fnmatch(file, '*.jpg'):
        miner_file.append(file)
```

```python
miner_dict = {'File': miner_file, 'Label': 'miner'}
df_miner = pd.DataFrame(miner_dict)
```

```python
file_path_miner= "/content/drive/MyDrive/EyeCoffee_TrainingData/coffee-leaves-disease/miner_img_xml"
df_miner["Filepath"] = file_path_miner + "/" + df_miner["File"]
df_miner.head()
```

## Combine

```python
from sklearn.utils import shuffle

frames = [df, df_rust, df_miner]
df_combined = pd.concat(frames)
df_combined = shuffle(df_combined, random_state=42)
df_combined
```

```python
df_combined['Label'].unique()
```

```python
df_combined['class_name'] = df_combined['Label'].map({'miner': 0, 'healthy': 1, 'rust': 2, 'red_spider_mite': 3})
```

```python
df_combined.rename(columns = {'Label':'class_name', 'class_name':'Label'}, inplace = True)
```

```python
df_combined['class_name'].unique()
```

```python
df_combined
```

```python
df_combined.to_csv("/content/drive/MyDrive/EyeCoffee_TrainingData/dataset.csv")
```
